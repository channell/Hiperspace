```polyglot-notebook
#r "nuget: Microsoft.EntityFrameworkCore"
#r "nuget: RocksDB"
#r "nuget: Hiperspace"
#r "nuget: Hiperspace.Rocks"
#r @"C:\Users\steve\source\repos\HiperspaceHub\examples\Hiperspace.Sparx\bin\Release\net8.0\Hiperspace.Sparx"
open Hiperspace.Rocks
open Sparx
open Microsoft.EntityFrameworkCore
open Sparx.EA
let rocks = @"C:\Hiperspace\Sparx"
let ssd = new RockSpace(rocks, MetaModel())
let space = new SparxSpace(ssd)
```

```polyglot-notebook
let javas = query {for e in space.Packages do 
                    where (e.Name = "java") 
                    select e}
            |> Seq.toList
```
